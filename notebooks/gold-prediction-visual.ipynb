```python
import matplotlib.pyplot as plt
import seaborn as sns

gold = spark.table("ddca26online23.default.player_pred_probs")

plt.figure(figsize=(16,6))
gold_df = gold.orderBy(gold['pred_prob'].desc())
sns.barplot(x='player_id', y='pred_prob', data=gold_df.limit(20).toPandas(), palette='coolwarm')
plt.xticks(rotation=90)
plt.ylabel('Predicted Win Probability')
plt.title('Top 20 Players Predicted Win Probability')
plt.show()
```
